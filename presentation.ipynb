```python
themes = {
    "Love & Romance": 32,
    "Peace & Unity": 8,
    "Empowerment / Identity": 10,
    "Nostalgia / Memory": 5,
    "Fantasy / Fairytale": 6,
    "Other / Novelty": 4
}

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(themes.values(),
        labels=themes.keys(),
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Set3.colors)
plt.title("Themes of Eurovision Winning Songs (1956–2025)", fontsize=14, weight="bold")
plt.show()

# Bar chart (alternative)
plt.figure(figsize=(10,6))
plt.bar(themes.keys(), themes.values(), color="orchid")
plt.xticks(rotation=30, ha="right")
plt.ylabel("Number of Winning Songs")
plt.title("Themes of Eurovision Winning Songs (1956–2025)", fontsize=14, weight="bold")
plt.show()
```
